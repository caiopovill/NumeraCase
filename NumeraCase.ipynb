```python
def barPerc(df,title,xVar,ax):
    '''
    barPerc(): Adicionar porcentagem de matizes aos gráficos de barras
    argumentos:
        df: pandas dataframe
        xVar: (string) X variable 
        ax: Axes object (para plotagem de contagem/barra Seaborn ou
                         enredo de barra de pandas)
    '''
    # 1. quantas categorias X
    ##   verifique se há NaN e remova
    numX=len([x for x in df[xVar].unique() if x==x])

    # 2. As barras são criadas em ordem de hue, organize-as
    bars = ax.patches
    ## 2a. Para cada variável X
    for ind in range(numX):
        ## 2b. Obtenha todas as hue barras
        hueBars=bars[ind:][::numX]
        ## 2c. Obtenha a altura total (para porcentagens)
        total = sum([x.get_height() for x in hueBars])

        # 3. Printa a % nas barras
        for bar in hueBars:
            ax.text(bar.get_x() + bar.get_width()/2.,
                    bar.get_height(),
                    f'{bar.get_height()/total:.0%}',
                    ha="center",va="bottom")
        ax.set(title=title)
```

# Separando os dados apenas em ofertas informativas

```python
oferta_informativa = df_final.loc[df_final['oferta']=='informativo',:]
info_social_email = oferta_informativa.loc[(oferta_informativa['canal'] == "['web', 'email', 'mobile']")]
info_web_email = oferta_informativa.loc[(oferta_informativa['canal'] == "['email', 'mobile', 'social']")]
```

Classificação entre oferta recebida ou visualizada para cada tipo de canal de venda e os diferentes gêneros

```python
todos = [('canal - social_email',info_social_email),('canal - web_email',info_web_email)]
fig = plt.figure(figsize = (18, 12))
axis = fig.add_axes([0, 0, 1, 1])
x = 1

for column in todos:
    plt.subplot(2, 2, x)
    ax = sns.countplot(x='genero',data=column[1], hue='tipo_evento')
    barPerc(column[1], column[0], 'genero', ax)
    x += 1
```

A diferença entre gênero quanto a visualização de ofertas informativas são insignificantes, mas o canal web_email se mostrou muito mais eficiente para levar a informação até o consumidor.
